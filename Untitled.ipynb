## Uncover the factors that lead to employee attrition and explore important questions such as:

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df.columns
```

```python
#PART ONE
sns.barplot(data = df, y = 'DistanceFromHome', x = 'JobRole', hue = 'Attrition')
plt.xticks(rotation=45)
```

```python
#PART TWO
sns.barplot(data = df, y = 'MonthlyIncome', x = 'Education', hue = 'Attrition')
```
